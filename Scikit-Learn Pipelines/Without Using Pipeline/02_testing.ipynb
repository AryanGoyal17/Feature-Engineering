```python
import pandas as pd
import pickle
import numpy as np
```

```python
model = pickle.load(open('models/decision_tree_model.pkl', 'rb'))
ohe_sex = pickle.load(open('models/ohe_sex.pkl', 'rb'))
ohe_emb = pickle.load(open('models/ohe_emb.pkl', 'rb'))
scalar_age = pickle.load(open('models/scalar_age.pkl', 'rb'))
scalar_fare = pickle.load(open('models/scalar_fare.pkl', 'rb'))
```

```python
test_input = np.array([1, 0, 0, 32, 30, 'S', 'male'], dtype=object).reshape(1, 7)
test_input
```

```python
test_input_sex = ohe_sex.transform(test_input[:, 6].reshape(1,1))
test_input_sex
```

```python
test_input_emb = ohe_emb.transform(test_input[:, 5].reshape(1,1))
test_input_emb
```

```python
test_input_age = scalar_age.transform(test_input[:, 3].reshape(1,1))
test_input_age
```

```python
test_input_fare = scalar_fare.transform(test_input[:, 4].reshape(1,1))
test_input_fare
```

```python
test_input_transformed = np.concatenate((test_input[:, 0:3], test_input_age, test_input_fare, test_input_emb, test_input_sex), axis=1)
test_input_transformed
```

```python
test_input_transformed.shape
```

```python
prediction = model.predict(test_input_transformed)
prediction
```
